```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
w1= pd.read_csv(r'D:\1\ALY6150\1\ARDI.csv')
w1.head(5)
```

```python
w1.describe(include='all')
```

```python
cause_counts = w1['Cause_of_Death'].value_counts()

top_10_causes = cause_counts[:10]

plt.figure(figsize=(10,6))
top_10_causes.plot(kind='bar')
plt.title('Top 10 Causes of Death')
plt.xlabel('Cause of Death')
plt.ylabel('Count')
plt.show()
```

```python
death_under_21 = w1[w1['AgeCategory'] == 'Under 21']
```

```python
cause_counts_under_21 = death_under_21['Cause_of_Death'].value_counts()

top_10_causes_under_21 = cause_counts_under_21[:10]

plt.figure(figsize=(10,6))
top_10_causes_under_21.plot(kind='bar')
plt.title('Top 10 Causes of Death for Under 21')
plt.xlabel('Cause of Death')
plt.ylabel('Count')
plt.show()
```
